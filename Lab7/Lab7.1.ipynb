## 1&2

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
file_path = '/Users/anantagarwal/Downloads/wine+quality/winequality-red.csv'
data = pd.read_csv(file_path, delimiter=';')
```

## 3

```python
X = data.drop('quality', axis=1)
y = data['quality']
```

## 4

```python
def categorize_quality(quality):
    if quality >= 7:
        return 'best'
    elif quality >= 5:
        return 'average'
    else:
        return 'poor'
```

```python
y = y.apply(categorize_quality)
```

## 5

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

## 6

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

## 7

```python
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
```

```python
y_pred = rf_model.predict(X_test)
```

## 8

```python
conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
```

## 9

```python
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
```

```python
print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
```

## 10

```python
rf_model_params = RandomForestClassifier(n_estimators=50, criterion='entropy', max_depth=10, 
                                         min_samples_split=3, min_samples_leaf=2, 
                                         max_features='sqrt', random_state=42)
```

```python
rf_model_params.fit(X_train, y_train)
```

```python
y_pred_params = rf_model_params.predict(X_test)
```

```python
accuracy_params = accuracy_score(y_test, y_pred_params)
precision_params = precision_score(y_test, y_pred_params, average='weighted')
recall_params = recall_score(y_test, y_pred_params, average='weighted')
f1_params = f1_score(y_test, y_pred_params, average='weighted')
```

```python
print(f"\nModified Parameters - Accuracy: {accuracy_params:.2f}")
print(f"Modified Parameters - Precision: {precision_params:.2f}")
print(f"Modified Parameters - Recall: {recall_params:.2f}")
print(f"Modified Parameters - F1-Score: {f1_params:.2f}")
```
